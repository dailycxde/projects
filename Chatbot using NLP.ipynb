```python
import nltk
from nltk.chat.eliza import Chat
import random
import string

# Tokenize sentences
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# Stemming
nltk.download('averaged_perceptron_tagger')
from nltk.stem import WordNetLemmatizer

# Create stopwords list
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# Prepare data
conversation_pairs = [
    [
        r"hey|hello",
        ["Hello!", "Hi there!"]
    ],
    [
        r"how are you?",
        ["I'm doing well, thank you.", "I'm good, how about you?"]
    ],
    [
        r"location|locations|places",
        ["Dubai Mall", 'Ferrari World', 'Expo City']
    ],
    [
        r"burj khalifa",
        [
            "The Burj Khalifa is the tallest skyscraper in the world, standing at a height of 828 meters (2,722 feet).",
            "It is located in Dubai, United Arab Emirates, and was officially opened on January 4, 2010.",
            "The building features a total of 163 floors and is a prominent symbol of Dubai's modern architecture and development."
        ]
    ],
    [
        r"history",
        ["The United Arab Emirates (UAE) was formed on December 2, 1971, bringing together seven emirates: Abu Dhabi, Dubai, Sharjah, Ras Al Khaimah, Ajman, Umm Al Quwain, and Fujairah. The union marked a pivotal moment in the region's history, leading to the establishment of a modern, prosperous nation known for its economic diversification and global influence."]
    ],
    [
        r"sorry (.*)",
        ["No problem, don't worry about it.", "It's alright."]
    ],
    [
        r"best hotels|good hotels|hotel options",
        [
            "Sure! Here are some popular hotels in Dubai: The Burj Al Arab, Atlantis The Palm, Jumeirah Beach Hotel.",
            "If you're looking for luxury, consider The Ritz-Carlton or The Address Downtown.",
            "For a more budget-friendly option, you might want to check out Novotel Dubai Al Barsha or Premier Inn Dubai International Airport."
        ]
    ],
    [
        r"quit",
        ["Goodbye!", "Nice chatting with you. See you soon!"]
    ]
]

# Create chatbot
chatbot = Chat(conversation_pairs)

# Run chatbot
print("Welcome to Chatbot! Type quit to exit.")
while True:
    try:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            break
        response = chatbot.respond(user_input.lower())
        print("Chatbot:", response)
    except (KeyboardInterrupt, EOFError, SystemExit):
        break
```
